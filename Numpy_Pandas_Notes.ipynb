## NumPy

Numpy handles homogeneous multidimensional arrays (ndarray class). Elements are indexed by a tuple of nonnegative integers.

**List vs NumPy**

* faster to read, less bites of memory, no type checking when iterating through objects
* Numpy uses fixed types (each character is by default treated as Int 32 (4 bytes)) while lists stores several information about a character (size, reference count, object type, object value) 
* NumPy uses contiguous memory, uses SIMD vector processing, effective cache utilization
* supports more operations on arrays

### PANDAS

Main objects are: **Series** (1d data structure i.e. column), **DataFrame** (2-d data structure i.e. table).


#### DIFFERENT FILE FORMATS
1. **CSV**

    **pd.read_csv(`<filepath/url>`, `<options>`)** reads a csv file given its filepath. 
    
    Some of the available options are: `index_col` (column that will be set for row indices), `na_values` (list of missing values), `parse_dates` (list of datetime columns), `date_parser=lambda x: pd.datetime.strptime(x,format=<timeformat>)` to indicate how dates should be converted to datetime format. 

    **`<dataframe>`.to_csv(`<filepath>`)** exports a data frame to csv file

2. **TSV** 
    **pd.read_csv(`<filepath/url>`, sep=`\t`)** reads a csv file given its filepath. 
    
    **`<dataframe>`.to_csv(`<filepath>.tsv`,sep=`\t`)** exports a data frame to csv file
3. **EXCEL** 
    
    Install necessary packages: xlwt (for writing to xls) openpyxl (for writing to xlsx) xlrd (for reading excel files)
    
    **pd.read_excel(`<filepath/url>`, `<options>`)** reads an excel file given its filepath. 
 
    **`<dataframe>`.to_excel(`<filepath>.xlsx`, `<options>`)** exports a data frame to an excel file 
4. **JSON**
    
    **pd.read_json(`<filepath/url>`, `<options>`)** reads a json file given its filepath. 
 
    **`<dataframe>`.to_json(`<filepath>.json`, `<options>`)** exports a data frame to a json file 
    
    Some options are: `orient` (e.g. 'records' etc) , `lines=True` (each row is in a different line). We must use the same options when reading the file. 
    
5. **DATABASES**

    Install necessary packages: SQLAlchemy (this is an ORM) psychopg2-binary (if you are using postgres):
    `from sqlalchemy import create_engine`
    `import psychopg2` 
    Create database connection: 
    `engine=create_engine(postgresql://dbuser:dbpass@localhost:5432/sample_db)` #use environment variables and config 
    
    **pd.read_sql(`table_name`, engine)** to import a table from a database
 
    **`<dataframe>`.to_sql(`table_name`, engine, if_exists='replace/append')** exports a data frame to export table to database 
    
    **pd.read_sql_query('SQL query string', engine)** 
    

#### DATA FRAME AND SERIES METHODS

 
1. **info** provides no. of entries (rows and columns) and the name of each column together with some info about the data type and the number of non-null entries
2. **pd.set_option('display.max_columns',`<int>`)** or **pd.set_option('display.max_rows',`<int>`)** will change the setting of how many columns/rows you can view in the notebook. 
3. **head(`<int>`)** and **tail(`<int>`)** to view only a specified number of rows at the beginning or end of the data frame. The argument is optional and defaults to 5. 
4. **DataFrame(`<table/dictionary>`)** creates a data frame from a table or dictionary of lists
6. **value_counts(normalize=False)** returns a list of distinct values in the column and the number of times they each appear in the column. Setting normalize=True will give percentages. 
7. **str.`<string method>`** can be applied to a series to implement a string method (returns another series, does not change the data frame unless reassigned)
8. **apply(`<function_name>`)** can be applied to a series to implement a function on each entry of the series (returns another series, does not change the data frame unless reassigned). Can be applied to data frames as well but returns series where function is either applied to each column (default) or each row (axis ='columns')
9. **applymap(`<function_name>`)** can be applied to a data frane to implement a function on each entry of the data frame (returns another data frame, does not change the data frame unless reassigned).
10. **map(`<dictionary>`)** and **replace(`<dictionary>`)** are series methods used to replace values of a series using a dictionary argument (like rename). For original values not found as a key in the dictionary, map will be replace them with NaN, while replace will keep the original values the same. 
11. **sort_values(by='col_name', ascending=True, inplace=True)** or **sort_values(by=col_list, ascending=bool_list,inplace=True)** to sort values based on a column or a list of columns. This can also be applied to a series without any arguments. 
12. **nlargest(`<int>`)** when used in a series, it displays the k (the integer argument) largest values in the series. Or **nlargest(`<int>`, 'col_name')** when used in a data frame, it displays all rows whose col_name values are the top k (integer argument) in the data 
13. **nsmallest** is the opposite of nlargest. 
14. **describe** gives statistics on numerical columns
15. **count** counts the non-missing values of a column
16. **groupby(column_list)** returns a DataFrameGroupby Object which groups the rows based on their common values in the given column list
17. **concat(series1,series2,axis='columns',sort=False)** will concatenate two series  
18. **isna** returns a data frame of boolean values to represent which entries are missing from the data frame
19. **fillna(value)** replace missing values with a string or a value
20. **astype(data_type)** casts entries to the given data type. Note that (np.nan) is a float and cannot be converted to int.
21. **unique** (series method) gives an array of unique values in a series
22. **pd.to_datetime(`<series>`, format=`<timeformat>`)**  convert a string to pd.datetime format using the following as reference https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
23. **dt.`<datetime method>`** can be applied to a series to implement a datetime method (returns another series, does not change the data frame unless reassigned)
24. **resample(Frequency_string)** resample is a time-based groupby, we can group the data by frequency code (one per day, one per week etc, see https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects). Resample is followed by a reduction/aggregate method (max,min etc). This returns a series/dataframe. 

#### DATA FRAME ACCESS/MANIPULATION
If `df` is a DataFrame object, 

1. `df[column_name]` will print the column of df labeled column_name
2. `df[col_list]` will print the columns of df whose labels appear in the list col_list
3. `df.iloc[<int>]` will print the row of df indexed by the integer (starts at 0)
3. `df.iloc[<int1>,<int2>]` will print the entry of df with row index int1 and column index int2
4. `df.iloc[list_int]` will print the rows of df indexed by the integers in the given list
5. `df.iloc[list2, list2]` will print the entries of df whose row index is in list1 and column index is in list2 
6. `df.loc[ind_list1,ind_list2]` same as iloc but the indices can be string labels. Note, we if we replace ind_list with a:b, then it will caputre rows/columns a to b (inclusive). An alternative is `df.at[ind1,ind2]` for accessing only one entry. 
7. `df.set_index('column_label',inplace=False)` will set the specified column entries as the row identifier. Does not modify data frame unless inplace is set to True. 
8. `df.reset_index(inplace=True)` resets the row indices to the default integers (starting with 0). 
9. `df.sort_index()` to sort the display alphabetically based on the index. 
10. `filt=df[col_name]==value` returns a series of boolean values for each row depending on whether the specified column is equal to the filter value or not. Running `df[filt]` or `df.loc[filt]` will only display the rows that have True in filt. Python builtin NOT, AND and OR will not work for Pandas. We need to use - (NOT), & (AND) and | (OR). 
11. `filt=df[col_name].isin(filt_list)` to filter only rows whose col_name value is in the given filt_list
12. `filt=df[col_name].str.contains(substring,na=False)` to filter only rows whose col_name contains a particular substring
13. `df.rename(columns={oldcol1_name : newcol1_name, oldcol3_name : new_col3_name}, inplace=True)` to rename specific columns
14. `df.drop(columns=[colname1, ..., colnamek],inplace=True)` to delete columns of the data frame whose names are as indicated in the list or `df.drop(index=k)` to drop a row or a list of rows
15. `df.dropna(axis='index', how='any')` drops rows (or columns if axis='columns') if any of its entries is missing (or if all entries are missing if how='all'). If we specify a subset argument, `df.dropna(axis='index', how='any',subset=col_list)` then it will only drop rows with missing values in that subset
15. `df.append(<series/dictionary/dataframe>, ignore_index=True,sort=False)` returns a data frame that adds rows/data frames to df

Other Functions: `min`,`max`,`mean`,`median`,`count`

#### DATA FRAME GROUPBY METHODS and RESAMPLING
If df_grp is a Data Frame Groupby object obtained by grouping data frame df using groupby, then 

1. `df_grp.get_group(value)` returns a data frame for the group corresponding to the given value
2. `df_grp['col_name].value_counts()` will return a series where a row is an aggregate value counts for a particular group
3. `df_grp['col_name].agg(['median','mean'])` will return a data frame where each row corresponds to a group and each column corresponds to an aggregate value 
4. `df.resample('W').agg({col1_name : 'mean', col2_name : 'max',..., colk_name='sum'})` resamples the time series weekly using different aggregate methods for each column
