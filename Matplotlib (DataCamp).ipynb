```python
#LINE PLOTS AND SCATTER PLOTS
fig, ax = plt.subplots(figsize=(5,2)) 

ax.plot(x,y,marker="o",linestyle=(0,(2,1)),color="r")
ax.set_title("Example Title")
ax.set_xlabel("Time (months)")
ax.set_ylabel("Random number (0-100)")
ax.set_xticklabels(x,rotation=90)
ax.tick_params('y',color='r',colors='r')
ax.annotate("Look here!", 
            xy=("Oct",1), 
            xytext=(7.75,22),
            arrowprops={"arrowstyle":"->", "color":"blue"})
ax.axvline(x="Jun",label="bmonth",linestyle="--")
ax.legend()
#KEYWORD ARGUMENT OPTIONS:
#marker: https://matplotlib.org/stable/api/markers_api.html
#linestyle: 
    #"" (no line)
    #(offset, (on_off_sequence)) 
    #"solid"/(0,())/"-" 
    #"dotted"/(0,(1,1))/":"
    #"dashed"/(0,(2,1))/"--" 
    #"dashdot"/(0,(2,1,1,1))/"-."


plt.show()
#display the interface
```

```python
#MULTIPLE LINE PLOTS IN ONE FIGURE

fig, ax = plt.subplots(figsize=(5,2)) 
ax.plot(x,y)
ax.plot(x,y2)

plt.show()
#display the interface
```

```python
#MULTIPLE LINE PLOTS IN ONE FIGURE

fig, ax = plt.subplots(2,1,figsize=(5,4),sharey=True)
#ax object is now a 1-by-2 array of axis objects
ax[0].plot(x,y,color='b')
ax[0].plot(x,y,marker='o',color='r',linestyle="")
ax[1].plot(x,y2,color='b')
ax[1].plot(x,y2,marker='o',color='r',linestyle="")
```

```python
fig, ax = plt.subplots(2,2,figsize=(10,4),sharey=True,sharex=True)
#ax object is now a 2-by-2 array of axis objects
ax[0,0].plot(x,y)
ax[0,1].plot(x,y2)
ax[1,0].plot(x,y3)
ax[1,1].plot(x,y4)
```

```python
fig,ax=plt.subplots(figsize=(5,2))
ax.plot(x,y,color='blue')
ax.set_xlabel("months")
ax.set_ylabel("Random Number (0-100)")
ax.tick_params('y',colors="blue")

ax2=ax.twinx()
ax2.plot(x,[10*i for i in y2],color='r')
ax2.set_ylabel("Random Number (0-100)x10")
ax2.tick_params('y',colors="r")
plt.show()
```

```python
#BAR CHARTS
fig,ax=plt.subplots(1,2,figsize=(10,2))
ax[0].bar(x,y,label="First")
ax[0].bar(x,y2,bottom=y,label="Second") #creates a stacked bar chart where 2nd values are on top of the first
ax[0].bar(x,y3,bottom=y+y2,label="Third") #creates a stacked bar chart where 2nd values are on top of the first
ax[0].legend()
ax[1].bar("First",y.mean(),yerr=y.std())
ax[1].bar("Second",y2.mean(),yerr=y2.std())
ax[1].bar("Third",y3.mean(),yerr=y3.std())
plt.show()
```

```python
#HISTOGRAMS
fig,ax=plt.subplots(figsize=(5,2))
ax.hist(y,label="First",bins=5)
ax.hist(y2,label="Second",bins=[0,7,15,29,40,53,71,99,100],histtype="step")
ax.legend()
plt.show()
```

# Data Communication Concepts

Making results stick: **Simple**, **Concrete**, **Credible**

Why data storytelling?
* *Anecdotes* = imagination
* *Stories* = memorable
* *adds value* provide context
* capture audience attention
* facilitate decision-making
* drive change


Data Storytelling Roadmap: Story -> Tech or Nontech? -> Data -> Viz -> Present 

* 3-minutes story
* Big idea
* Must be **insightful**, **explanatory**, **concise**

Your audience
1. What do they know? Adjust content. 
2. What do they need to understand? Be conversaional.
3. What level of information do they need? Serve the audience. 
* Be receptive to questions
* Proactively ensure understanding
* Explain differently

**A.D.E.P.T** -> Analogy, Diagram, Example, Plain English, Technical definition

Narrative Structure
1. Background - What motivated the analysis? What changes prompted the analysis? Who is the focus of the analysis?
2. Insight - Only relevant information and supporting evidence about what contributed to the problem
3. Climax/Central Insight - What would happen if there is no change?
4. Next Steps - potential solutions/course of action/proactive

Building Narrative
1. Change of factors over time
2. Correlation of variables
3. Comparison of groups
4. Clustering 

McCandless method (Visualization)
1. Introduce visualization by name (Graph Headline must be clear and obvious, use y by x technique)
2. Anticipate audience's questions
3. State insights
4. Help the audience relate (importance, action items)

Written Reports 
**TYPES** 

1. Informational - Contains only factual info, short, not strict structure, inform about facts
2. Analytical Report - contains analysis (relationships/recommendations), length varies, follows strict structure, informs data-driven decisions
3. Final Report - contains data analysis, findings and results, visuals, usually long and meant for audience who require a lot of detail
4. Summary Report - contains key findings and recommendations and visuals, short (<5 pages), summary of final report, may contain link to main document, meant for audience who don't require technical detail

**STRUCTURE**
1. Introduction - contains sections on purpose, contextual information and questions of analysis
2. Body - contains sections on Data, Method, Analysis and Visuals, Results
3. Conclusions - restate question and summarize important results

**BEST PRACTICES**
* 1-3-25 Rule (1 page abstract, 3-page executive summary, 25-page detail) 
* Keep Track of how results were produced
    * Reproducibility - getting the same results if we run the same analysis 
    * Replicability - getting the similar results if we follow same approach
* Avoid Manual data manipulation
* Document randomness
* Make code interpretable 
* Cite Bibliography correctly
* Be concise and precise
* avoid misleading and confusing phrases
* avoid empty phrases
* use concrete nouns
* avoid run-on sentences 

Oral Presentations 
**CONSIDERATIONS FOR PLANNING**
1. Purpose - informative or instructional or persuasive?
2. Audience - technical team or executive team or customer? small meeting or conference or large meeting or workshop?
3. Message - opening message, central statement, closing statement

**STRUCTURE**
1. Introduction
2. Methods/Analysis/Model Outputs
3. Conclusions and Takeaways

**BEST PRACTICES**
* PLAN: plan your talk
* CONTENT: One message per slide 
* COLOR: use least amount of colors needed (no more than 3 colors), consider color deficiencies
* FONT: sans-serif is conventionally used for slides while serif is used for print
* HEADLINE: Use headlines that highlight the main point
* LAYERING AND HIGHLIGHTING
* VISUALIZATIONS
* PRACTICE
* Engage the audience
* Leave time for Q&A
